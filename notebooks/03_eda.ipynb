```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("../data/features_data.csv")
```

```python
RESULTS_DIR = "../results"
os.makedirs(RESULTS_DIR, exist_ok=True)
```

```python
df["lifesat"].hist(bins=20)
plt.title("Distribution of Life Satisfaction")
plt.xlabel("Life Satisfaction")
plt.ylabel("Frequency")
plt.savefig(f"{RESULTS_DIR}/lifesat_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.scatter(df["social_trust"], df["lifesat"])
plt.xlabel("Social Trust")
plt.ylabel("Life Satisfaction")
plt.title("Life Satisfaction vs Social Trust")
plt.savefig(f"{RESULTS_DIR}/lifesat_vs_trust.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.scatter(df["work_stress"], df["lifesat"])
plt.xlabel("Work Stress")
plt.ylabel("Life Satisfaction")
plt.title("Life Satisfaction vs Work Stress")
plt.savefig(f"{RESULTS_DIR}/lifesat_vs_workstress.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.scatter(df["income"], df["lifesat"])
plt.xlabel("Income")
plt.ylabel("Life Satisfaction")
plt.title("Life Satisfaction vs Income")
plt.savefig(f"{RESULTS_DIR}/lifesat_vs_income.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df.groupby("gender")["lifesat"].mean().plot(kind="bar")
plt.title("Life Satisfaction by Gender")
plt.savefig(f"{RESULTS_DIR}/lifesat_vs_gender.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.scatter(df["age"], df["lifesat"])
plt.title("Life Satisfaction by Age")
plt.savefig(f"{RESULTS_DIR}/life_sat_by_age.png", dpi=300, bbox_inches="tight")
plt.show()
```
