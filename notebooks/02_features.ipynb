```python
import pandas as pd
import os

```

```python
df = pd.read_csv("../data/cleaned_data.csv")
```

```python
RESULTS_DIR = "../results"
os.makedirs(RESULTS_DIR, exist_ok=True)
```

```python
trust_cols = [
    "trustpeople", "trustfair", "trusthelp", "trustneig"
]

df["social_trust"] = df[trust_cols].mean(axis=1)
```

```python
dis_cols = [col for col in df.columns if col.startswith("dis")]

df["discrimination_index"] = df[dis_cols].mean(axis=1)
```

```python
work_stress_cols = [
    "workstress", "jobstress1", "worktired",
    "workdifhome", "workdiffam", "workdiffre"
]

existing_cols = [col for col in work_stress_cols if col in df.columns]

df["work_stress"] = df[existing_cols].mean(axis=1)
```

```python
relig_cols = [
    "religious", "pray", "salat", "quran"
]

existing_cols = [col for col in relig_cols if col in df.columns]

df["religiosity"] = df[existing_cols].mean(axis=1)
```

```python
social_cols = [
    "meetfri", "meetnei", "meetrela", "meetcomp"
]

df["social_activity"] = df[social_cols].mean(axis=1)
```

```python
econ_cols = [
    "satinc", "satjob1", "satwork"
]

existing_cols = [col for col in econ_cols if col in df.columns]

df["economic_satisfaction"] = df[existing_cols].mean(axis=1)
```

```python
well_cols = [
    "welhappy", "welenjoy", "welworry", "welsad"
]

existing_cols = [col for col in well_cols if col in df.columns]

df["wellbeing"] = df[existing_cols].mean(axis=1)
```

```python
df = df[[
    "lifesat",
    "age",
    "gender",
    "income",
    "social_trust",
    "discrimination_index",
    "work_stress",
    "religiosity",
    "social_activity",
    "economic_satisfaction",
    "wellbeing"
]]
```

```python
df.to_csv("../data/features_data.csv", index=False)
```
